```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Personal/VIT/Big Data Frameworks/BDA J component/data/By_caste/Unemployment_Rate_By_Caste_All_India/c1.csv")
```

```python
x1 = df.Month
y1 = df.Upper_caste

plt.plot(x1, y1, label = "Upper caste")

y2 = df.Intermediate
plt.plot(x1, y2, label = 'Intermediate')

y3 = df.OBC
plt.plot(x1, y3, label = 'OBC' )

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate based on caste')
plt.legend()
plt.show()
```

```python
x1 = df.Month
y1 = df.SC

plt.plot(x1, y1, label = "SC")

y2 = df.ST
plt.plot(x1, y2, label = 'ST')

y3 = df.Caste_not_avail
plt.plot(x1, y3, label = 'Caste not available' )

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate based on caste')
plt.legend()
plt.show()
```
