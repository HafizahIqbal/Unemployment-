```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Personal/VIT/Big Data Frameworks/BDA J component/data/naukri_jobspeaksurvey/main.csv")
```

```python
plt.plot(df.Month, df.Overall, color='red', marker='o')
plt.title('Change in employment Rate in all industries', fontsize=14)
plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)
plt.xlabel('Month')
plt.ylabel('Employment Rate')
plt.show()
```

```python
x1 = df.Month
y1 = df.Banking_and_financial_services

plt.plot(x1, y1, label = "Banking and financial services")

y2 = df.Healthcare
plt.plot(x1, y2, label = 'Health care')

y3 = df.Education
plt.plot(x1, y3, label = 'Education' )

y4 = df.Export_import
plt.plot(x1, y4, label = 'Export and import')

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in employment rate in different industries')
plt.legend()
plt.show()
```
