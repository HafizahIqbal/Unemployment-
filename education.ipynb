```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Personal/VIT/Big Data Frameworks/BDA J component/data/by_education/c1.csv")
```

```python
x1 = df.Month
y1 = df.No_education

plt.plot(x1, y1, label = "No education")

y2 = df.Primary_upto_5th_std
plt.plot(x1, y2, label = 'Primary (upto 5th std)')

y3 = df.Secondary_6th_9th_std
plt.plot(x1, y3, label = 'Secondary (6th-9th std)' )

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate based on education')
plt.legend()
plt.show()
```

```python
x1 = df.Month
y1 = df.Higher_secondary_10th_12th_std

plt.plot(x1, y1, label = "Higher secondary (10th-12th std)")

y2 = df.gt_eq_Graduate
plt.plot(x1, y2, label = '> = Graduate')

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate based on education')
plt.legend()
plt.show()
```
