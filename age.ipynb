```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Personal/VIT/Big Data Frameworks/BDA J component/data/by_age_group/Unemployment_Rate_By_Age_Group_All_India/c1.csv")
```

```python
x1 = df.Month
y1 = df.a15_19years

plt.plot(x1, y1, label = "15-19 years")

y2 = df.a20_24years
plt.plot(x1, y2, label = '20-24 years')

y3 = df.a25_29years
plt.plot(x1, y3, label = '25-29 years' )

y4 = df.a30_34years
plt.plot(x1, y4, label = '30-34 years')

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate age wise')
plt.legend()
plt.show()
```
