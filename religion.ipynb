```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Personal/VIT/Big Data Frameworks/BDA J component/data/By_religion/c1.csv")
```

```python
plt.style.use('dark_background')
x1 = df.Month
y1 = df.Hindu

plt.plot(x1, y1, label = "Hindu")

y2 = df.Muslim
plt.plot(x1, y2, label = 'Muslim')

y3 = df.Christian
plt.plot(x1, y3, label = 'Christian' )

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate based on religion')
plt.legend()
plt.show()
```
