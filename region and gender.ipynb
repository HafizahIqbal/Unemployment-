```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Personal/VIT/Big Data Frameworks/BDA J component/data/by_region_gender/c1.csv")
```

```python
x1 = df.Month
y1 = df.All_India_total

plt.plot(x1, y1, label = "All India total")

y2 = df.All_India_male
plt.plot(x1, y2, label = 'All India male')

y3 = df.All_India_female
plt.plot(x1, y3, label = 'All_India_female' )

plt.rcParams["figure.figsize"] = (20,10)
plt.xticks(rotation = 90)

plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Changes in Unemployment rate overall India based on gender')
plt.legend()
plt.show()
```
